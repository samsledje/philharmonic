```python
import re

import pandas as pd

ROOT_DIR = Path(
    "/mnt/home/ssledzieski/Projects/philharmonic-experiments/results/recipe_dream/20250110-recipe/max_20_prots_added"
    # "/mnt/home/ssledzieski/Projects/philharmonic-experiments/results/recipe_dream/new_download"
)


def map_size(x):
    if x < 16:
        return "Small"
    elif x < 32:
        return "Medium"
    else:
        return "Large"


map_lr = {
    "linear_0.1": "ReCIPE-10%",
    "linear_0.25": "ReCIPE-25%",
    "sqrt": "ReCIPE-Sqrt",
    "pre-recipe": "Control",
}

map_dream = {
    "d1": "DREAM1",
    "d2": "DREAM2",
    "d3": "DREAM3",
}

column_map = {
    "jaccard_score": "Jaccard Score",
    "f1_score_by_pval": "F1 Score",
    "cluster_size": "Cluster Size",
}

df_list = []
for i in sorted(ROOT_DIR.glob("*recipe*")):
    print(i)
    dream_net, lin_ratio = re.match(r".*(d\d)_recipe_(.*).csv", str(i)).groups()
    df_i = pd.read_csv(i)
    df_i["dream"] = map_dream[dream_net]
    df_i["linear_ratio"] = map_lr[lin_ratio]
    df_i["cluster_size"] = df_i["num_proteins"].apply(map_size)
    df_list.append(df_i)
    if "Unnamed: 0" in df_i.columns:
        df_i.drop(columns=["Unnamed: 0"], inplace=True)
results_df = pd.concat(df_list)
results_df = results_df.sort_values(by=["dream", "linear_ratio"])
results_df.rename(columns=column_map, inplace=True)
```

## vs. Randomized

```python
import pandas as pd

ROOT_DIR = Path(
    "/mnt/home/ssledzieski/Projects/philharmonic-experiments/results/recipe_dream/20250110-recipe/max_20_prots_added"
    # "/mnt/home/ssledzieski/Projects/philharmonic-experiments/results/recipe_dream/20250110-recipe/no_max_added"
    # "/mnt/home/ssledzieski/Projects/philharmonic-experiments/results/recipe_dream/new_download"
)
```

```python
import numpy as np
from scipy import stats


def test_method_vs_random(method_scores, random_scores):
    """
    Test if method scores are significantly higher than random baseline.

    Parameters:
    method_scores: array-like of shape (k,), where k is number of samples
        The scores from your method
    random_scores: array-like of shape (k, 50)
        For each sample, the 50 scores from random baseline

    Returns:
    dict containing:
        - p_value: significance of test
        - percentile_ranks: percentile rank of each method score
        - median_percentile: median percentile rank across samples
    """
    # Ensure inputs are numpy arrays
    method_scores = np.array(method_scores)
    random_scores = np.array(random_scores)

    k = len(method_scores)
    percentile_ranks = []

    # For each sample
    for i in range(k):
        # Combine method score with random scores
        all_scores = np.append(random_scores[i], method_scores[i])

        # Calculate percentile rank of method score
        # (number of scores it exceeds + 0.5 * number of scores it equals) / total number
        rank = np.sum(all_scores < method_scores[i])
        rank += 0.5 * np.sum(all_scores == method_scores[i])
        percentile_rank = rank / len(all_scores)

        percentile_ranks.append(percentile_rank)

    # Convert to array
    percentile_ranks = np.array(percentile_ranks)

    # Perform one-sided Wilcoxon signed-rank test
    # H0: median percentile rank = 0.5
    # H1: median percentile rank > 0.5
    statistic, p_value = stats.wilcoxon(
        percentile_ranks - 0.5,  # center around 0 for null hypothesis
        alternative="greater",  # one-sided test
    )

    return {
        "p_value": p_value,
        "percentile_ranks": percentile_ranks,
        "median_percentile": np.median(percentile_ranks),
    }
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 3, figsize=(18, 15))

metric = "jaccard_score"
# metric = "f1_score_by_count"

linear_ratio = "linear_0.1"

dream_result_ids = {"linear_0.1": (3, 0), "linear_0.25": (4, 1), "sqrt": (6, 2)}

for dci, dream_challenge in enumerate(["d1", "d2", "d3"]):
    dream_results = sorted(
        [i for i in ROOT_DIR.iterdir() if (i.name.startswith(dream_challenge))]
    )
    # for i in dream_results:
    #     print(i)

    base_results = pd.read_csv(dream_results[dream_result_ids[linear_ratio][0]])
    base_results["Setting"] = "ReCIPE"
    base_results.sort_values(by="cluster_id")
    rand_results = pd.read_csv(dream_results[dream_result_ids[linear_ratio][1]])
    rand_results = rand_results[rand_results["cluster_id"] != "cluster_id"]
    rand_results["Setting"] = "Random"
    rand_results.sort_values(by="cluster_id")
    concat_results = pd.concat([base_results, rand_results])

    method_scores = base_results[metric]

    # First verify we have 50 scores per cluster
    score_counts = rand_results.groupby("cluster_id").size()
    if not all(score_counts == 50):
        raise ValueError("Not all clusters have exactly 50 scores")

    # Get unique clusters in order
    clusters = rand_results["cluster_id"].unique()
    k = len(clusters)

    # Initialize the output array
    random_scores = np.zeros((k, 50))

    # Fill the matrix
    for i, cluster in enumerate(clusters):
        random_scores[i] = rand_results[rand_results["cluster_id"] == cluster][
            metric
        ].values

    metric_title = metric.replace("_", " ").title()
    dream_title = f"DREAM {dream_challenge[1]}"
    plot_df = pd.DataFrame(
        {
            f"Mean Random {metric_title}": rand_results.groupby("cluster_id")[
                metric
            ].mean(),
            f"ReCIPE {metric_title}": base_results.groupby("cluster_id")[metric].mean(),
        }
    )

    result_stats = test_method_vs_random(method_scores, random_scores)

    sns.scatterplot(
        plot_df,
        x=f"Mean Random {metric_title}",
        y=f"ReCIPE {metric_title}",
        s=15,
        ax=ax[0][dci],
    )
    sns.boxplot(
        concat_results, x="Setting", y=metric, ax=ax[1][dci], order=["Random", "ReCIPE"]
    )
    # ax[0][dci].set_xscale('log')
    # ax[0][dci].set_yscale('log')
    # ax[1][dci].set_yscale('log')
    ax[1][dci].set_ylabel(metric_title)
    ax[0][dci].set_title(f"{dream_title} per Cluster")
    ax[1][dci].set_title(
        f"{dream_title} Aggregate \n p = {result_stats['p_value']:.2e}"
    )
    # plt.suptitle(f"{dream_challenge} {metric_title}")

plt.savefig(
    IMG_DIR / f"recipe_{linear_ratio}_{metric}_random_bootstraps.svg",
    bbox_inches="tight",
    dpi=300,
)
plt.savefig(
    IMG_DIR / f"recipe_{linear_ratio}_{metric}_random_bootstraps.png",
    bbox_inches="tight",
    dpi=300,
)
plt.show()
```
